```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('compressed_data.csv')
df
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['last review']= pd.to_datetime(df['last review'], errors ='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review': df['last review'].min()},inplace=True)
```

```python
df.dropna(subset =['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=["license","house_rules"],errors='coerce')
```

```python
df.head()
```

```python
#emove dollar signs and convert to float
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head(2)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

Visualization

What is the distribution of listing prices?


```python
df.head(2)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title("Distribution of listing prices")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

How are diff room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='room type',data=df,color='hotpink')
plt.title("Room type distributions")
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

How are listing distributions across different neighborhoods? 

```python
plt.figure(figsize=(8,5))
sns.countplot(y='neighbourhood group',data=df,color='blue',order=df['neighbourhood group'])
plt.title("Number of listing by neighbourhood group")
plt.xlabel('Count')
plt.ylabel('Neighburhood Group')
plt.show()
```

Reviews Over Time

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of reviews over time')
plt.xlabel("Date")
plt.ylabel('Number of reviews')
plt.show()
```

Price VS Room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type',data = df,palette='Set1')
plt.title("Price vs Room Type")
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```
